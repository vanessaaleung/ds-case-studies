```
import pandas as pd
import numpy as np
import math
import statistics
from google.colab import files
```

The data is already filtered to only 25 movies.

```
df = pd.read_csv("https://raw.githubusercontent.com/vanessaaleung/rawdata/master/HW4_Data.xlsx%20-%20Filtered%20Data.csv")
df = df.fillna(0)
df.set_index('Serial Number', inplace=True)
df.head()
```

# Original Item-Item

```
all_rating = []
for col in range(len(df.columns)):
  target = df.iloc[:, col].values
  df2 = df.drop(df.columns[col], axis=1)
  df3 = df2.as_matrix()
  cos = df3 * target.reshape((99,1))
  numerator = cos.sum(axis = 0)
  deno1 = np.repeat((math.sqrt(sum(map(lambda i : i * i, target)))), 24)
  deno2 = []
  for col in df2:
    deno2.append(math.sqrt(sum(map(lambda i : i * i, df2[col]))))
  cosine = numerator/(deno1 * deno2)
  rating = []
  for i in range(len(df2)):
    rating.append(np.sum(cosine * df2.iloc[i, :])/sum(cosine))

  all_rating.append(rating)
```

```
original_item = pd.DataFrame(np.transpose(all_rating))
header = list(df)
original_item.columns = header
original_item.head()
```

# Item-Item Mean Centered

```
df4 = pd.read_csv("https://raw.githubusercontent.com/vanessaaleung/rawdata/master/HW4_Data.xlsx%20-%20Filtered%20Data.csv")
df4.set_index('Serial Number', inplace=True)
```

```
average = df4.mean(axis = 0)
df5 = df4
for col in range(len(df5.columns)):
  newdata = []
  for i in range(len(df5.iloc[:, col])):
    if df5.iloc[i, col] > 0:
      df5.iloc[i, col] = df5.iloc[i, col] - average[col]
```

```
all_rating2 = []
for col in range(len(df5.columns)):
  target = df5.iloc[:, col].values
  target[np.isnan(target)] = 0
  df6 = df5.drop(df5.columns[col], axis=1)
  df6 = df6.fillna(0)
  df7 = df6.as_matrix()
  cos = df7 * target.reshape((99,1))
  numerator = cos.sum(axis = 0)
  deno1 = np.repeat((math.sqrt(sum(map(lambda i : i * i, target)))), 24)
  deno2 = []
  for j in df6:
    deno2.append(math.sqrt(sum(map(lambda i : i * i, df6[j]))))
  cosine = numerator/(deno1 * deno2)
  abscosine = abs(cosine)
  rating = []
  for i in range(len(df6)):
    rating.append(average[col] + np.sum(cosine * df6.iloc[i, :])/sum(abscosine))
  all_rating2.append(rating)
```

```
item_mean = pd.DataFrame(np.transpose(all_rating2))
header = list(df)
item_mean.columns = header
item_mean.head()
```

# Item-Item Zscored

```
df8 = pd.read_csv("https://raw.githubusercontent.com/vanessaaleung/rawdata/master/HW4_Data.xlsx%20-%20Filtered%20Data.csv")
df8.set_index('Serial Number', inplace=True)
```

```
average = df8.mean(axis = 0)
stdev = []
for i in range(len(df8.columns)):
  stdev.append(statistics.stdev(df8.iloc[:, i].dropna()))
df9 = df8
for col in range(len(df9.columns)):
  newdata = []
  for i in range(len(df9.iloc[:, col])):
    if df9.iloc[i, col] > 0:
      df9.iloc[i, col] = df9.iloc[i, col] - average[col]/stdev[col]
```

```
all_rating3 = []
for col in range(len(df9.columns)):
  target = df9.iloc[:, col].values
  target[np.isnan(target)] = 0
  df10 = df9.drop(df9.columns[col], axis=1)
  df10 = df10.fillna(0)
  df11 = df10.as_matrix()
  cos = df11 * target.reshape((99,1))
  numerator = cos.sum(axis = 0)
  deno1 = np.repeat((math.sqrt(sum(map(lambda i : i * i, target)))), 24)
  deno2 = []
  for j in df10:
    deno2.append(math.sqrt(sum(map(lambda i : i * i, df10[j]))))
  cosine = numerator/(deno1 * deno2)
  abscosine = abs(cosine)
  rating = []
  for i in range(len(df10)):
    rating.append(stdev[col] * np.sum(cosine * df10.iloc[i, :])/sum(abscosine) + average[col])
  all_rating3.append(rating)
```

```
item_zscored = pd.DataFrame(np.transpose(all_rating3))
header = list(df8)
item_zscored.columns = header
item_zscored.head()
```
